```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.utils.vis_utils import plot_model
from yellowbrick.classifier import ConfusionMatrix
```

```python
base = datasets.load_iris()

print("Shape =", base.data.shape)

print("\nFeature Names")
[print(i) for i in base.feature_names]

print("\nTarget Names")
[print(i) for i in base.target_names]

print("\nTarget\n", base.target)
```

```python
previsores = base.data
classe = base.target

print(previsores.shape, classe.shape)
```

```python
classe_dummy = np_utils.to_categorical(classe) # Transformando as classes de iris em colunas
print(classe_dummy)
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores, classe_dummy, test_size=0.3, random_state=0)

print("\nDisposição da divisão Treino/Teste\n")

print("X Treino:", x_train.shape[0], "registros x", x_train.shape[1], "variáveis preditoras")
print("Y Treino:", y_train.shape[0], "registros x", y_train.shape[1], "variáveis de resposta\n")

print("X Teste:", x_test.shape[0], "registros x", x_test.shape[1], "variáveis preditoras")
print("Y Teste:", y_test.shape[0], "registros x", y_test.shape[1], "variáveis de resposta\n")
```

```python
modelo = Sequential()
modelo.add(Dense(units=5, input_dim=4)) #input_dim = variáveis preditoras
modelo.add(Dense(units=4))
modelo.add(Dense(units=3, activation="softmax")) #units = classe_dummy | softmax = Mais de duas classes na classificação
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"]) 
```

```python
modelo.fit(x_train, y_train, epochs=1000, validation_data=(x_test, y_test))
```

```python
previsoes = modelo.predict(x_test)
previsoes
```

```python
previsoes = (previsoes>0.5)
previsoes
```

```python
y_test_matriz = [np.argmax(i) for i in y_test]
print(np.array(y_test_matriz))
```

```python
y_previsao_matriz = [np.argmax(i) for i in previsoes]
print(np.array(y_previsao_matriz))
```

```python
matriz = confusion_matrix(y_test_matriz, y_previsao_matriz)
matriz
```
